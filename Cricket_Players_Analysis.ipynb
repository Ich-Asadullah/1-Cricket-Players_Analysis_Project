#### Q2: TODO - Who has Played most Matches in All Formats Combined

````python
df["Total_Matches"] = df["BATTING_T20s_Mat"] + df["BATTING_ODIs_Mat"] + df["BATTING_Tests_Mat"]
most_mat =  df.sort_values(by="Total_Matches", ascending=False).head(10)

most_mat[["NAME", "COUNTRY", "Total_Matches"]]
````

````python
top_10_players = most_mat[["NAME", "COUNTRY", "BATTING_T20s_Mat", "BATTING_ODIs_Mat", "BATTING_Tests_Mat"]]

ax = top_10_players.plot(kind="bar", x="NAME", stacked=True, figsize=(12, 6), title = ("Players with most Experience"))

plt.legend(title="Match Type", labels=["ODI", "T20I", "Test"])
````

#### Q3: TODO - Which player has Taken Most Wickets in all Formats Combined

````python
df["Total_Wickets"] = df["BOWLING_T20s_Wkts"] + df["BOWLING_ODIs_Wkts"] + df["BOWLING_Tests_Wkts"]

most_wkts =  df.sort_values(by="Total_Wickets", ascending=False).head(10)

most_wkts = most_wkts[["NAME", "COUNTRY", "Total_Wickets"]]

most_wkts
````

````python
top_10_bowlers = most_mat[["NAME", "COUNTRY", "BOWLING_T20s_Wkts", "BOWLING_ODIs_Wkts", "BOWLING_Tests_Wkts"]]

ax = top_10_bowlers.plot(kind="bar", x="NAME", stacked=True, figsize=(12, 6), title = ("Players with most Runs"))

plt.legend(title="Match Type", labels=["ODI", "T20I", "Test"])
````

````python
heatmap_data = top_10_bowlers[["BOWLING_T20s_Wkts", "BOWLING_ODIs_Wkts", "BOWLING_Tests_Wkts"]]

heatmap_data.index = top_10_bowlers["NAME"]

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", linewidths=.5)

plt.ylabel("Player Name")
plt.title("Bowling Wickets Heatmap")
````
